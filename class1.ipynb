```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels import regression
```

```python
RiskFree = 0.01
RiskFreeDaily = RiskFree / 252
SP5 = pd.read_excel('data.xlsx', sheet_name=1)[['Date', 'Adj Close']].set_index('Date')
AMZ = pd.read_excel('data.xlsx', sheet_name=2)[['Date', 'Adj Close']].set_index('Date')

print(len(SP5))
print(SP5.columns)
print(SP5)

SP5Returns = SP5.pct_change().dropna() - RiskFreeDaily
AMZReturns = AMZ.pct_change().dropna() - RiskFreeDaily
```

```python
''' R_asset = Alpha + Beta_asset *  R_benchmark + e_i '''

def linreg(x,y):
    x = sm.add_constant(x)
    model = regression.linear_model.OLS(y,x).fit()
    return model.params[0], model.params[1]

alpha, beta = linreg(SP5Returns, AMZReturns)
print([alpha, beta])

```

```python
''' Weekly '''
SP5ReturnsW = SP5.pct_change(periods = 7).dropna() - RiskFreeDaily
AMZReturnsW = AMZ.pct_change(periods = 7).dropna() - RiskFreeDaily

alphaW, betaW = linreg(SP5ReturnsW, AMZReturnsW)
print([alphaW, betaW])

''' Monthly '''
SP5ReturnsM = SP5.pct_change(periods = 21).dropna() - RiskFreeDaily
AMZReturnsM = AMZ.pct_change(periods = 21).dropna() - RiskFreeDaily

alphaM, betaM = linreg(SP5ReturnsM, AMZReturnsM)
print([alphaM, betaM])
```

```python
''' Alternative data range: 5 last years '''
SP5Returns5 = SP5.pct_change().dropna().loc['2012':'2017'] - RiskFreeDaily
AMZReturns5 = AMZ.pct_change().dropna().loc['2012':'2017'] - RiskFreeDaily

alpha5, beta5 = linreg(SP5Returns5, AMZReturns5)
print([alpha5, beta5])
```
